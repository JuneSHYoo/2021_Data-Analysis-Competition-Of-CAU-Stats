```python
unique_pum = [
    '배추', '무', '양파', '건고추','마늘',
    '대파', '얼갈이배추', '양배추', '깻잎',
    '시금치', '미나리', '당근',
    '파프리카', '새송이', '팽이버섯', '토마토',
]

unique_kind = [
    '청상추', '백다다기', '애호박', '캠벨얼리', '샤인마스캇'
]
```

```python
uniq = unique_pum + unique_kind
```

```python
import statsmodels.graphics.tsaplots as sgt 
import matplotlib.pyplot as plt
for pum in uniq:
    tmp = data[data['품목'] == pum]
    y_train = tmp[tmp['1_week'] > 0][['1_week']][:-28]


        ## Close
    fig = plt.figure()
    print(pum)
    ax2 = fig.add_subplot(1,1,1)

    sgt.plot_pacf(y_train, lags = 30, zero = False, method = ('ols'), ax=ax2)
    ax2.set_title("PACF")

    plt.show()
```
