## 기후 lag

```python
lag = data[['date','요일','품목','주산지','temp','max_Temp','min_Temp','wind','rain']]
lag.품목.unique()
```

```python
lag.주산지.unique()
```

```python
lag
```

### temp_lag

```python
lag['t_lag_1'] = lag.groupby('품목')['temp'].shift(1)   #8일전
lag['t_lag_2'] = lag.groupby('품목')['temp'].shift(2)   #9일전
lag['t_lag_3'] = lag.groupby('품목')['temp'].shift(3)   #10일전
lag['t_lag_4'] = lag.groupby('품목')['temp'].shift(4)   #11일전
lag['t_lag_2w'] = lag.groupby('품목')['temp'].shift(7)  #14일전
lag['t_lag_3w'] = lag.groupby('품목')['temp'].shift(14) #21일전
lag['t_lag_1m'] = lag.groupby('품목')['temp'].shift(23) #한달전
lag['t_lag_2m'] = lag.groupby('품목')['temp'].shift(53) #두달전
```

```python
lag['t_lead_1'] = lag.groupby('품목')['temp'].shift(-6)  # 1_week 기준 하루전
lag['t_lead_2'] = lag.groupby('품목')['temp'].shift(-5)  # 2일전
lag['t_lead_3'] = lag.groupby('품목')['temp'].shift(-4)  # 3일전
lag['t_lead_4'] = lag.groupby('품목')['temp'].shift(-3)  # 4일전
```

```python
lag['max_t_lead_1'] = lag.groupby('품목')['max_Temp'].shift(-6)  # 1_week 기준 하루전
lag['max_t_lead_2'] = lag.groupby('품목')['max_Temp'].shift(-5)  # 2일전
lag['max_t_lead_3'] = lag.groupby('품목')['max_Temp'].shift(-4)  # 3일전
lag['max_t_lead_4'] = lag.groupby('품목')['max_Temp'].shift(-3)  # 4일전
```

```python
lag = lag.fillna(-1)
```

```python
df = pd.merge(data, lag , on =['date','요일','품목','주산지','temp','max_Temp','min_Temp','wind','rain'] , how = 'left')
df
```
